```python
import pandas as pd

file_path = r'C:\my-project\my-project\my-project\DM_JOB\Query_Check\Raw_data.xlsx'
df = pd.read_excel(file_path)
df.head()
```

```python
# 공통 열 지정
common_cols = df.columns[:11].to_list()
all_cols = df.columns
extra_cols = all_cols[11:]

# 'CRF_VER_NM_'으로 시작하는 열 찾기
split_points = [i for i, col in enumerate(all_cols) if str(col).startswith('CRF_VER_NM_')]

# 공통 열 지정 (처음 11개 열)
common_cols = df.columns[:11].to_list()
all_cols = df.columns
extra_cols = all_cols[11:]

# 'CRF_VER_NM_'으로 시작하는 열 위치 찾기
split_points = [i for i, col in enumerate(all_cols) if str(col).startswith('CRF_VER_NM_')]

# 구간 나누기
sheet_ranges = []
for i in range(len(split_points)):
    start = split_points[i]
    end = split_points[i + 1] if i + 1 < len(split_points) else len(all_cols)  # 여기 수정됨
    sheet_ranges.append((start, end))

# 시트별 저장
with pd.ExcelWriter(r'C:\my-project\my-project\my-project\DM_JOB\Query_Check\Cleaning_RawData.xlsx', engine='openpyxl') as writer:
    for idx, (start, end) in enumerate(sheet_ranges):
        range_cols = list(all_cols[start:end])
        selected_cols = common_cols + range_cols

        df_sheet = df[selected_cols].copy()

        # 'CRF_STAT_'으로 시작하는 열 중 하나라도 'C' 값이 있는 행만 필터링
        crf_stat_cols = [col for col in range_cols if str(col).startswith('CRF_STAT_')]

        if crf_stat_cols:
            mask = df_sheet[crf_stat_cols].eq('C').any(axis=1)
            df_sheet = df_sheet[mask]

        sheet_name = f'raw_{idx + 1}'
        df_sheet.to_excel(writer, sheet_name=sheet_name, index=False)
```
