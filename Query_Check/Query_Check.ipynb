```python
import pandas as pd

# 엑셀 파일 경로
file_path = "BETA_C19184_재발행.xlsx"

# 모든 시트 이름 확인
excel_file = pd.ExcelFile(file_path)
sheet_names = excel_file.sheet_names # 시트명 리스트에 담기
```

```python
# for문으로 시트명 하나씩 확인하기
for i, sheet_name in enumerate(sheet_names, start = 1): 
    print(f"{i}번째 시트 이름: {sheet_name}")
```

- enumerate는 리스트에서 하나씩 꺼내면서 인덱스(번호)도 같이 불러옴.
- enumerate(start= 1)로 디폴트 인덱스 0부터지만 1부터 시작하도록 지정.
- f-string를 활용하여 문자열 안에 변수를 {}로 바로 넣어서 출력 가능.

```python
# PL_df와 RD_df 병합(key: 'SUBJ_ID', PL 기준 left join)
# 병합할 때 중복되는 열은 컬럼명 뒤에 '_x', '_y' 붙여서 구분
PL_merged_df = pd.merge(PL_df, RD_df, how='left', on = 'SUBJ_ID', suffixes=('', '_RD'))
PL_merged_df.head()
```

```python
# 날짜형 변환
PL_merged_df['CRF_INSDT'] = pd.to_datetime(PL_merged_df['CRF_INSDT'], errors='coerce')
PL_merged_df['RANDOMDT'] = pd.to_datetime(PL_merged_df['RANDOMDT'], errors='coerce')
```

```python
# CRF 등록일시, 랜덤일 날짜 차이 계산
PL_merged_df['dt_diff'] = (PL_merged_df['CRF_INSDT'] - PL_merged_df['RANDOMDT']).dt.days

# 구분 열 추가(음수면 '확인필요', 아니면 '확인불필요')
PL_merged_df['diff_check'] = PL_merged_df['dt_diff'].apply(lambda x: '확인필요' if x < 0 else '확인불필요')
PL_merged_df['diff_check'].value_counts()
```

```python
# 'diff_check' 값이 '확인필요'인 대상자 중에서 RDSTGEN와 PLSEX 값 동일한지 확인
PL_merged_df[PL_merged_df['diff_check'] == '확인필요'][['SUBJ_ID', 'RDSTGEN', 'PLSEX']].head(10)   
# 구분 열 추가(동일하면 '동일', 아니면 '불일치')
PL_merged_df['SEX_check'] = PL_merged_df.apply(lambda x: '동일' if x['RDSTGEN'] == x['PLSEX'] else '불일치', axis=1)
PL_merged_df['SEX_check'].value_counts()
```
